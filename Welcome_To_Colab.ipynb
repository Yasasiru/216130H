```
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from google.colab import files

# Set display options
pd.set_option("display.float_format", "{:.2f}".format)

# Define your stock tickers and time range
tickers = ['ABNB', 'NVDA', 'WMT']
start_date = '2019-04-01'
end_date = '2025-04-01'
```

```
# Download data for each ticker
df_ABNB = yf.download('ABNB', start=start_date, end=end_date)
df_NVDA = yf.download('NVDA', start=start_date, end=end_date)
df_WMT  = yf.download('WMT',  start=start_date, end=end_date)

# Flatten MultiIndex columns (in case they exist)
for df in [df_ABNB, df_NVDA, df_WMT]:
    df.columns = [col[0] if isinstance(col, tuple) else col for col in df.columns]

# Store cleaned DataFrames
dfs = {
    'ABNB': df_ABNB[['Open', 'High', 'Low', 'Close', 'Volume']].dropna(),
    'NVDA': df_NVDA[['Open', 'High', 'Low', 'Close', 'Volume']].dropna(),
    'WMT':  df_WMT[['Open', 'High', 'Low', 'Close', 'Volume']].dropna()
}

# Optional: Save and download CSVs
for ticker, df in dfs.items():
    df.to_csv(f"{ticker}_data.csv")
    files.download(f"{ticker}_data.csv")
```

```
scaler = MinMaxScaler()
scaled_data = {}

for ticker, df in dfs.items():
    print(f"\n--- {ticker} Summary ---")
    display(df.describe())

    # Plot Closing Price
    plt.figure(figsize=(10, 4))
    plt.plot(df['Close'], label='Close Price', color='blue')
    plt.title(f'{ticker} - Closing Price Over Time')
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.show()

    # Box Plot
    plt.figure(figsize=(6, 5))
    sns.boxplot(y=df['Close'], color='skyblue')
    plt.title(f'{ticker} - Box Plot of Closing Price')
    plt.ylabel("Price")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    # Correlation Heatmap
    plt.figure(figsize=(6, 4))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    plt.title(f'{ticker} - Feature Correlation')
    plt.show()

    # Scale and store
    scaled_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns, index=df.index)
    scaled_data[ticker] = scaled_df
```

```
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(sequence_length, len(data)):
        X.append(data[i-sequence_length:i])
        y.append(data[i])
    return np.array(X), np.array(y)
```

```
sequence_length = 60
epochs = 20
batch_size = 32

results = {}

for ticker, df in scaled_data.items():
    print(f"\n📊 Training LSTM for {ticker}...")

    # Focus only on Close prices
    close_prices = df['Close'].values.reshape(-1, 1)

    # Create sequences
    X, y = create_sequences(close_prices, sequence_length)
    X = X.reshape((X.shape[0], X.shape[1], 1))  # 3D for LSTM

    # Train-test split
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Define LSTM model
    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
        Dropout(0.2),
        LSTM(50, return_sequences=False),
        Dropout(0.2),
        Dense(1)
    ])

    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)

    # Predict
    y_pred = model.predict(X_test)

    # Inverse scale
    scale_min = df['Close'].min()
    scale_max = df['Close'].max()
    y_test_unscaled = y_test * (scale_max - scale_min) + scale_min
    y_pred_unscaled = y_pred.flatten() * (scale_max - scale_min) + scale_min

    # Evaluate
    rmse = np.sqrt(mean_squared_error(y_test_unscaled, y_pred_unscaled))
    print(f"✅ {ticker} RMSE: {rmse:.2f}")

    # Save results
    results[ticker] = {
        'model': model,
        'rmse': rmse,
        'y_test': y_test_unscaled,
        'y_pred': y_pred_unscaled
    }

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(y_test_unscaled, label='Actual', color='blue')
    plt.plot(y_pred_unscaled, label='Predicted', color='orange')
    plt.title(f'{ticker} - LSTM Prediction vs Actual')
    plt.xlabel('Time Step')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```

```
print("\n📈 Model Performance Summary:")
for ticker, res in results.items():
    print(f"{ticker}: RMSE = {res['rmse']:.2f}")
```

```
from itertools import product

# Tuning options
param_grid = {
    'sequence_length': [30, 60],
    'lstm_units': [50, 100],
    'dropout_rate': [0.2, 0.3],
    'batch_size': [32],
    'epochs': [10]
}

# Create all combinations
param_combinations = list(product(
    param_grid['sequence_length'],
    param_grid['lstm_units'],
    param_grid['dropout_rate'],
    param_grid['batch_size'],
    param_grid['epochs']
))

# Function to create LSTM model
def build_lstm_model(input_shape, units, dropout_rate):
    model = Sequential([
        LSTM(units, return_sequences=True, input_shape=input_shape),
        Dropout(dropout_rate),
        LSTM(units, return_sequences=False),
        Dropout(dropout_rate),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```
ticker = 'ABNB'
df = scaled_data[ticker]
close_prices = df['Close'].values.reshape(-1, 1)

best_rmse = float('inf')
best_params = None
best_model = None

for (seq_len, units, dropout, batch_size, epochs) in param_combinations:
    # Create sequences
    X, y = create_sequences(close_prices, seq_len)
    X = X.reshape((X.shape[0], X.shape[1], 1))

    # Split
    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Build and train
    model = build_lstm_model((X_train.shape[1], 1), units, dropout)
    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)

    # Predict and evaluate
    y_pred = model.predict(X_test)
    scale_min = df['Close'].min()
    scale_max = df['Close'].max()
    y_test_unscaled = y_test * (scale_max - scale_min) + scale_min
    y_pred_unscaled = y_pred.flatten() * (scale_max - scale_min) + scale_min
    rmse = np.sqrt(mean_squared_error(y_test_unscaled, y_pred_unscaled))

    print(f"[{ticker}] Seq={seq_len}, Units={units}, Dropout={dropout} → RMSE: {rmse:.2f}")

    if rmse < best_rmse:
        best_rmse = rmse
        best_params = (seq_len, units, dropout, batch_size, epochs)
        best_model = model

print(f"\n✅ Best RMSE for {ticker}: {best_rmse:.2f}")
print(f"Best Params: Sequence={best_params[0]}, Units={best_params[1]}, Dropout={best_params[2]}")
```

```
def forecast_next_days(model, recent_data, n_days, scaler_min, scaler_max, sequence_length):
    forecast = []
    input_seq = recent_data[-sequence_length:].reshape(1, sequence_length, 1)

    for _ in range(n_days):
        pred = model.predict(input_seq, verbose=0)[0][0]
        forecast.append(pred)

        # Append prediction and roll window
        input_seq = np.append(input_seq[:, 1:, :], [[[pred]]], axis=1)

    # Inverse transform the forecast
    forecast_unscaled = np.array(forecast) * (scaler_max - scaler_min) + scaler_min
    return forecast_unscaled
```

```
forecast_horizon = 30
forecast_results = {}

for ticker, df in scaled_data.items():
    print(f"\n🔮 Forecasting 30 days for {ticker}...")

    # Get best model (assume you stored it from tuning)
    model = results[ticker]['model']

    # Data
    recent_data = df['Close'].values
    sequence_length = 60  # or best from tuning

    scale_min = dfs[ticker]['Close'].min()
    scale_max = dfs[ticker]['Close'].max()

    # Forecast
    forecasted_prices = forecast_next_days(model, recent_data, forecast_horizon, scale_min, scale_max, sequence_length)

    forecast_results[ticker] = forecasted_prices

    # Plot
    plt.figure(figsize=(10, 4))
    plt.plot(range(1, forecast_horizon+1), forecasted_prices, marker='o', linestyle='-')
    plt.title(f"{ticker} - Forecasted Close Prices (Next 30 Days)")
    plt.xlabel("Day")
    plt.ylabel("Forecasted Price")
    plt.grid(True)
    plt.show()
```
